```python
#convert categorical columns to dummy variables using get_dummies
house_df = pd.get_dummies(house_df, columns = catcol)
house_df.head()
```

```python
house_df.shape
```

```python
#create indepedent and dependent variables
x = house_df.drop(['SalePrice'], axis=1)
y = house_df['SalePrice']

from sklearn.model_selection import train_test_split
#split into 80/20 train/test datasets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

#create linear regression model
model = LinearRegression()
#train the model
model.fit(x_train, y_train)
#use model to make prediction on test set
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

#report R2 value and RMSE
print('RMSE:', sqrt(mean_squared_error(y_test, y_pred)))
print('R2:', r2_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
#assign standardscaler
ss = StandardScaler()
#fit and transform x training data
features_Xtrain = ss.fit_transform(x_train)
#create PCA that retains 90% of the variance
pca = PCA(n_components = 0.9, whiten = True)
#Conduct PCA
features_xtrain_pca = pca.fit_transform(features_Xtrain)
```

```python
#print original # and Reduced # of features
print('Original # of Features:', features_Xtrain.shape[1])
print('Reduced # of Features:', features_xtrain_pca.shape[1])
```

```python
#Transform but not fit the test features
features_xtest = ss.transform(x_test)
#conduct pca
features_xtest_pca = pca.transform(features_xtest)
```

```python
#create second linear regression for PCA data
model_two = LinearRegression()
#train model on pca data
model_two.fit(features_xtrain_pca, y_train)
#use model to make predictions on pca test set
y_pred_pca = model_two.predict(features_xtest_pca)
```

```python
#print RMSE and R2 for PCA data model
print('RMSE:', sqrt(mean_squared_error(y_test, y_pred_pca)))
print('R2:', r2_score(y_test, y_pred_pca))
```

```python
from sklearn import preprocessing
#create minmax scale
minmax_scale = preprocessing.MinMaxScaler(feature_range = (0,1))

#scale the feature
scaled_features = minmax_scale.fit_transform(x_train)
```

```python
#create high variance thresholder
from sklearn.feature_selection import VarianceThreshold

thresholder = VarianceThreshold(threshold = 0.1)
```

```python
#create high variance matrix
features_high_variance = thresholder.fit_transform(scaled_features)
```

```python
#view matrix
features_high_variance[0:3]
```

```python
#scale the feature
test_scaled_features = minmax_scale.transform(x_test)
```

```python
#create feature matrix and view matrix
test_features_high_variance = thresholder.transform(test_scaled_features)
test_features_high_variance[0:3]
```

```python
#create linear regression model
model_three = LinearRegression()
#train the model on high variance data
model_three.fit(features_high_variance, y_train)
#use model to get predictions on test set
y_pred_minmaxscale = model_three.predict(test_features_high_variance)
```

```python
#print out RMSE and R2
print('RMSE:', sqrt(mean_squared_error(y_test, y_pred_minmaxscale)))
print('R2:', r2_score(y_test, y_pred_minmaxscale))
```
